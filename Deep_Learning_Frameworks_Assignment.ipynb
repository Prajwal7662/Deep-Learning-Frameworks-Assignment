##Deep Learning Frameworks



1. What is TensorFlow 2.0, and how is it different from TensorFlow 1.x?
- TensorFlow 2.0 is the updated version of TensorFlow that focuses on simplicity, ease of use, and faster development.
Differences:

Eager Execution ON by default (runs line-by-line like Python).

Unified Keras API for building models easily.

No complex sessions or placeholders.

Cleaner, pythonic syntax.

Better performance and GPU utilization.

2. How do you install TensorFlow 2.0?

(Or: pip install tensorflow==2.x for a specific version)

3. What is the primary function of tf.function in TensorFlow 2.0?

- tf.function converts a Python function into a high-performance graph function.
It makes code run faster by building a computational graph.

4. What is the purpose of the Model class in TensorFlow 2.0?

- tf.keras.Model is the base class to create custom neural network architectures.
It allows you to:

define layers

write call() for forward pass

compile, train, evaluate models

5. How do you create a neural network using TensorFlow 2.0?

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

6. What is the importance of Tensor Space in TensorFlow?

- Tensor Space is the multi-dimensional data structure that TensorFlow uses.
Importance:

Holds all input/output data

Supports GPU operations

Enables automatic differentiation

7. How can TensorBoard be integrated with TensorFlow 2.0?

- Import TensorBoard callback:

tensorboard_cb = tf.keras.callbacks.TensorBoard(log_dir='./logs')


Use it in training:

model.fit(x, y, callbacks=[tensorboard_cb])


Run TensorBoard:

tensorboard --logdir=logs

```python
# 4. How can you visualize the training progress using TensorFlow and Matplotlib?
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# ---- Create dummy training data ----
x_train = np.random.rand(100, 1)      # 100 samples, 1 feature
y_train = 3 * x_train + 2 + np.random.randn(100, 1) * 0.1  # simple linear pattern

# ---- Build a simple model ----
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# ---- Train model ----
history = model.fit(x_train, y_train, epochs=20, verbose=0)

# ---- Visualize training loss ----
plt.plot(history.history['loss'])
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
plt.show()
```

```python
# 6. How do you create a simple neural network in PyTorch?
import torch
import torch.nn as nn

class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(10, 16)
        self.fc2 = nn.Linear(16, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

model = Net()
print(model)
```

```python
# 7. How do you define a loss function and optimizer in PyTorch?
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
# 8. How do you implement a custom loss function in PyTorch?
# Example: Mean Absolute Error (MAE)

import torch.nn as nn
import torch

class CustomMAE(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, y_pred, y_true):
        return torch.mean(torch.abs(y_pred - y_true))

loss_fn = CustomMAE()
```

```python
# 9. How do you save and load a TensorFlow model?

import tensorflow as tf
import numpy as np

# ---- Build model using Input() to avoid warnings ----
model = tf.keras.Sequential([
    tf.keras.Input(shape=(3,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# ---- Dummy data ----
x = np.random.rand(20, 3)
y = np.random.rand(20, 1)

# ---- Train ----
model.fit(x, y, epochs=5, verbose=0)

# ---- SAVE model using .keras extension ----
model.save("my_tf_model.keras")   # NEW correct format
print("Model saved successfully!")

# ---- LOAD model ----
loaded_model = tf.keras.models.load_model("my_tf_model.keras")
print("Model loaded successfully!")

# ---- Test loaded model ----
print(loaded_model.predict(x[:2]))
```
